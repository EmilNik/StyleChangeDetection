```python
import sys
import pandas as pd
from sklearn.preprocessing import StandardScaler

sys.path.append("..")
```

```python
train = pd.read_csv('../data/500_ngrams/train_features.csv', index_col=0)
val = pd.read_csv('../data/500_ngrams/validation_features.csv', index_col=0)
```

```python
train.head()
```

```python
number_of_features = int((train.shape[1] - 1) / 2)
number_of_features
```

```python
a_columns = ['A_{}'.format(i) for i in range(number_of_features)]
b_columns = ['B_{}'.format(i) for i in range(number_of_features)]

train_A = train[a_columns]
train_B = train[b_columns]

val_A = val[a_columns]
val_B = val[b_columns]
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

standard_scaler = StandardScaler()
min_max_scaler = MinMaxScaler(feature_range=(0, 1))

standard_scaler.fit(train_A)
min_max_scaler.fit(train_A)
```

```python
# Standard Scaler

train_scaled = pd.concat([pd.DataFrame(standard_scaler.transform(train_A), columns=a_columns),
                          pd.DataFrame(standard_scaler.transform(train_B), columns=b_columns),
                          train['different_author']],
                          axis=1)

val_scaled = pd.concat([pd.DataFrame(standard_scaler.transform(val_A), columns=a_columns),
                        pd.DataFrame(standard_scaler.transform(val_B), columns=b_columns),
                        val['different_author']],
                        axis=1)

# MinMax Scaler

train_minmax_scaled = pd.concat([pd.DataFrame(min_max_scaler.transform(train_A), columns=a_columns),
                                 pd.DataFrame(min_max_scaler.transform(train_B), columns=b_columns),
                                 train['different_author']],
                                 axis=1)

val_minmax_scaled = pd.concat([pd.DataFrame(min_max_scaler.transform(val_A), columns=a_columns),
                               pd.DataFrame(min_max_scaler.transform(val_B), columns=b_columns),
                               val['different_author']],
                               axis=1)
```

```python
train_scaled.to_csv('../data/500_ngrams/train_features_scaled_standard.csv')
val_scaled.to_csv('../data/500_ngrams/validation_features_scaled_standard.csv')
```

```python
train_minmax_scaled.to_csv('../data/500_ngrams/train_features_scaled_minmax.csv')
val_minmax_scaled.to_csv('../data/500_ngrams/validation_features_scaled_minmax.csv')
```
