```python
import sys
import pandas as pd
from sklearn.preprocessing import StandardScaler

sys.path.append("..")
```

```python
train = pd.read_csv('../data/500_ngrams/train_features.csv', index_col=0)

train['A'] = train['A'].apply(lambda x: pd.eval(x))
train['B'] = train['B'].apply(lambda x: pd.eval(x))

val = pd.read_csv('../data/500_ngrams/validation_features.csv', index_col=0)

val['A'] = val['A'].apply(lambda x: pd.eval(x))
val['B'] = val['B'].apply(lambda x: pd.eval(x))
```

```python
train.head()
```

```python
number_of_features = len((train['A'][0]))
number_of_features
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = StandardScaler()
min_max_scaler = MinMaxScaler(feature_range=(0, 1))

train_a_vectors = pd.DataFrame(train['A'].tolist())
train_b_vectors = pd.DataFrame(train['B'].tolist())
val_a_vectors = pd.DataFrame(val['A'].tolist())
val_b_vectors = pd.DataFrame(val['B'].tolist())

scaler.fit(train_a_vectors)
min_max_scaler.fit(train_a_vectors)

# transform train df
# train['A'] = scaler.transform(train_a_vectors).tolist()
# train['B'] = scaler.transform(train_b_vectors).tolist()


# # transform validation df
# val_a_vectors = pd.DataFrame(val['A'].tolist())
# val_b_vectors = pd.DataFrame(val['B'].tolist())

# val['A'] = scaler.transform(val_a_vectors).tolist()
# val['B'] = scaler.transform(val_b_vectors).tolist()
```

```python
a_columns_names = ['A_{}'.format(i) for i in range(833)]
b_columns_names = ['B_{}'.format(i) for i in range(833)]

# Standard Scaler

train_scaled = pd.concat([pd.DataFrame(scaler.transform(train_a_vectors), columns=a_columns_names),
                          pd.DataFrame(scaler.transform(train_b_vectors), columns=b_columns_names),
                          train['different_author']],
                          axis=1)

val_scaled = pd.concat([pd.DataFrame(scaler.transform(val_a_vectors), columns=a_columns_names),
                        pd.DataFrame(scaler.transform(val_a_vectors), columns=b_columns_names),
                        val['different_author']],
                        axis=1)

# MinMax Scaler

train_minmax_scaled = pd.concat([pd.DataFrame(min_max_scaler.transform(train_a_vectors), columns=a_columns_names),
                                 pd.DataFrame(min_max_scaler.transform(train_b_vectors), columns=b_columns_names),
                                 train['different_author']],
                                 axis=1)

val_minmax_scaled = pd.concat([pd.DataFrame(min_max_scaler.transform(val_a_vectors), columns=a_columns_names),
                               pd.DataFrame(min_max_scaler.transform(val_a_vectors), columns=b_columns_names),
                               val['different_author']],
                               axis=1)
```
